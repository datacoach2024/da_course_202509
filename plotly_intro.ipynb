```python
import pandas as pd
import plotly
import plotly.express as px
```

```python
tips = plotly.data.tips()
```

```python
tips.sample(10)
```

```python
px.histogram(
    data_frame=tips,
    x='tip'
)
```

```python
px.box(
    data_frame=tips,
    x='tip'
)
```

```python
tips.head(10)
```

```python
tips_by_sex = tips.groupby(['sex'], as_index=False).agg(
    avg_tip = ('tip', 'mean')
)
tips_by_sex
```

```python
px.bar(
    data_frame=tips_by_sex,
    x='sex',
    y='avg_tip'
)
```

```python
px.scatter(
    data_frame=tips,
    y='tip',
    x='total_bill',
    color_continuous_midpoint=.5,
)
```

```python
gpm = plotly.data.gapminder()
gpm.sample(10)
```

```python
gpm_life_exp = gpm.groupby(['year', 'continent'], as_index=False).agg(avg_life_exp=('lifeExp', 'mean'))
gpm_life_exp
```

```python
fig = px.line(
    data_frame=gpm_life_exp,
    x='year',
    y='avg_life_exp',
    color='continent'
)
fig.update_layout({
    'legend':{'orientation':'h', 'y':1.2},
    'xaxis':{'title':{'text':'observation_year', 'font':{'size':20}}, 'tickfont':{'size':16, 'color':'green'}}
})
```
