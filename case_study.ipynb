```python
with open('dicts/columns.json') as f:
    columns_dict = json.load(f)
```

```python
columns_dict['customers']
```

```python
def make_snake_case(col_name: str) -> str:
    new_col_name = ''
    for ix, sym in enumerate(col_name):
        if sym.isupper() and ix != 0:
            new_col_name += f"_{sym.lower()}"
        else:
            new_col_name += sym.lower()
    
    return new_col_name
```

```python
def read_data(sheet_name: str) -> pd.DataFrame:
    table_name = make_snake_case(sheet_name)
    usecols = columns_dict[table_name]

    df = pd.read_excel(
        SOURCE, 
        sheet_name=sheet_name,
    )
    df.rename(columns={col: make_snake_case(col) for col in df.columns}, inplace=True)

    return df[usecols]
```

```python
def load_data(df: pd.DataFrame, table_name: str) -> None:
    with set_connection() as pg:
        pg.execute(text(f"truncate table {table_name}"))
        pg.commit()

        df.to_sql(
            name=table_name,
            con=pg,
            index=False,
            if_exists='append'
        )
```

```python
load_data(df=customers2, table_name='customers')
```
